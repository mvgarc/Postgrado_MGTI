```python
import numpy as np
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import OLSInfluence

# Crear DataFrame
df = pd.DataFrame(data_wine_quality_from_text)

# Separar X e y
X = df.drop(columns='y_Quality')
y = df['y_Quality']

# Añadir constante (intercepto)
X_const = sm.add_constant(X)

# Regresión lineal múltiple con OLS
modelo_ols = sm.OLS(y, X_const).fit()
print(modelo_ols.summary())
```

```python
# Stepwise usando sklearn (solo 'forward' permitido)
selector = SequentialFeatureSelector(LinearRegression(), direction='forward', n_features_to_select='auto')
selector.fit(X, y)
features_selected = X.columns[selector.get_support()]
print("\nVariables seleccionadas por Stepwise:", list(features_selected))

# Modelo reducido con variables seleccionadas
X_stepwise = sm.add_constant(X[features_selected])
modelo_stepwise = sm.OLS(y, X_stepwise).fit()
print("\nResumen modelo Stepwise:")
print(modelo_stepwise.summary())
```

```python
# Contraste con Cp de Mallows
rss_p = np.sum(modelo_stepwise.resid ** 2)
rss_full = np.sum(modelo_ols.resid ** 2)
sigma2_full = rss_full / (X_const.shape[0] - X_const.shape[1])
p = X_stepwise.shape[1]  # incluye intercepto
n = X_const.shape[0]
cp = rss_p / sigma2_full - (n - 2 * p)
print(f"\nCp de Mallows: {cp:.3f} (ideal si está cerca de {p})")
```
