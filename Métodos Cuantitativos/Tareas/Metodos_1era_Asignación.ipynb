# **Influencia de la ausencia de indicadores de gestión en la toma de decisiones estratégicas en una empresa minorista de electrodomésticos**

La empresa minorista de electrodomésticos presenta deficiencias en la toma de decisiones estratégicas debido a la ausencia de indicadores de gestión. Esto genera incertidumbre en la planificación y control, afectando la eficiencia, productividad y competitividad del negocio. Sin indicadores, los directivos no pueden evaluar resultados ni alinear actividades con los objetivos, lo que reduce la capacidad de respuesta y deteriora el rendimiento general de la empresa.

**Variable independiente:** Ausencia de indicadores de gestión  
**Variable dependiente:** Toma de decisiones estratégicas  

**Indicadores seleccionados:**

- Frecuencia de reportes
- Uso de indicadores
- Tiempo de toma de decisiones
- Percepción de calidad en las decisiones

**Instrumento simulado:** Entrevistas estructuradas con 10 gerentes.


```python
# Importar librerías
import pandas as pd

# Crear los datos simulados
data = {
    'Nombre': [f'Gerente {i+1}' for i in range(10)],
    'Frecuencia_Reportes': ['mensual', 'semanal', 'mensual', 'nunca', 'semanal',
                            'mensual', 'nunca', 'mensual', 'semanal', 'semanal'],
    'Uso_Indicadores': ['bajo', 'medio', 'medio', 'bajo', 'alto',
                        'medio', 'bajo', 'medio', 'alto', 'medio'],
    'Tiempo_Toma_Decisiones': [5, 2, 4, 8, 1, 3, 7, 4, 2, 3],  # en días
    'Percepcion_Calidad': ['regular', 'buena', 'buena', 'mala', 'excelente',
                           'buena', 'mala', 'regular', 'buena', 'buena']
}
df = pd.DataFrame(data)
df
```

**Hipótesis nula (H₀):** No existe relación significativa entre la ausencia de indicadores de gestión y la toma de decisiones estratégicas en la empresa minorista.

**Hipótesis alternativa (H₁):** Existe una relación significativa entre la ausencia de indicadores de gestión y la toma de decisiones estratégicas en la empresa minorista.

**## Análisis de correlación (prueba paramétrica: Pearson)**

Como complemento del análisis no paramétrico, se aplica la prueba de correlación de Pearson. Aunque la variable "Uso de Indicadores" es ordinal, se codifica como escala numérica para fines exploratorios. Esto permite verificar si existe una correlación lineal entre el nivel de uso de indicadores y el tiempo de toma de decisiones.

```python
from scipy.stats import pearsonr

# Correlación de Pearson entre uso de indicadores (codificado) y tiempo de decisión
corr_pearson, p_pearson = pearsonr(df['Uso_Indicadores_Num'], df['Tiempo_Toma_Decisiones'])

print("Coeficiente de correlación de Pearson:", round(corr_pearson, 2))
print("Valor p:", round(p_pearson, 4))
```

**Interpretación del coeficiente de Pearson**

El coeficiente obtenido indica el grado de relación lineal entre las variables. Si el valor p es menor a 0.05, se rechaza la hipótesis nula y se confirma que existe una relación significativa.

- Un coeficiente negativo sugiere que a mayor uso de indicadores, menor tiempo para tomar decisiones.
- Este resultado se debe interpretar con cautela, ya que los datos no siguen distribución normal y provienen de una escala ordinal.


```python
from scipy.stats import f_oneway, kruskal, spearmanr

# Grupos según uso de indicadores
grupo_bajo = df[df['Uso_Indicadores'] == 'bajo']['Tiempo_Toma_Decisiones']
grupo_medio = df[df['Uso_Indicadores'] == 'medio']['Tiempo_Toma_Decisiones']
grupo_alto = df[df['Uso_Indicadores'] == 'alto']['Tiempo_Toma_Decisiones']

# 1. ANOVA (Prueba paramétrica para comparar medias entre grupos)
f_stat, p_val = f_oneway(grupo_bajo, grupo_medio, grupo_alto)

print("Análisis de Varianza (ANOVA)")
print("----------------------------")
print("Estadístico F:", round(f_stat, 3))
print("Valor p:", round(p_val, 4))
if p_val < 0.05:
    print("Resultado: Se rechaza la hipótesis nula.")
    print("Conclusión: Hay diferencias significativas en el tiempo de toma de decisiones según el nivel de uso de indicadores.\n")
else:
    print("Resultado: No se rechaza la hipótesis nula.")
    print("Conclusión: No hay diferencias significativas en el tiempo de toma de decisiones entre los grupos.\n")

# 2. Prueba no paramétrica Kruskal-Wallis (alternativa a ANOVA cuando no se cumple normalidad)
h_stat, p_kruskal = kruskal(grupo_bajo, grupo_medio, grupo_alto)

print("Prueba no paramétrica Kruskal-Wallis")
print("------------------------------------")
print("Estadístico H:", round(h_stat, 3))
print("Valor p:", round(p_kruskal, 4))
if p_kruskal < 0.05:
    print("Resultado: Se rechaza la hipótesis nula.")
    print("Conclusión: Existen diferencias significativas entre los grupos en el tiempo de toma de decisiones.\n")
else:
    print("Resultado: No se rechaza la hipótesis nula.")
    print("Conclusión: No existen diferencias significativas entre los grupos.\n")

# 3. Correlación no paramétrica de Spearman (para variables ordinales o no normales)
corr_spearman, p_spearman = spearmanr(df['Uso_Indicadores_Num'], df['Tiempo_Toma_Decisiones'])

print("Correlación de Spearman")
print("-----------------------")
print("Coeficiente rho:", round(corr_spearman, 3))
print("Valor p:", round(p_spearman, 4))
if p_spearman < 0.05:
    print("Resultado: Se rechaza la hipótesis nula.")
    print("Conclusión: Existe correlación significativa entre uso de indicadores y tiempo de toma de decisiones.")
else:
    print("Resultado: No se rechaza la hipótesis nula.")
    print("Conclusión: No hay correlación significativa entre las variables.")
```

Se realizaron distintas pruebas de hipótesis para validar la relación entre el uso de indicadores de gestión y el tiempo de toma de decisiones.

El **ANOVA** indicó que existen diferencias significativas en el tiempo de decisión entre los grupos bajo, medio y alto uso de indicadores (p < 0.05).

La **prueba no paramétrica Kruskal-Wallis** confirmó esta diferencia, apoyando los resultados de ANOVA en caso de que los datos no sigan distribución normal.

La **correlación no paramétrica de Spearman mostró** una asociación significativa negativa entre el nivel de uso de indicadores y el tiempo de decisión, reforzando que un mayor uso se asocia con tiempos más cortos para decidir.

**Analisis Final:** A partir del análisis de los datos simulados, se evidencia que la frecuencia de los reportes de desempeño varía entre semanal, mensual y la ausencia total de reportes, siendo la periodicidad mensual la más común entre los gerentes. Este hallazgo sugiere que no existe un sistema estandarizado ni constante de seguimiento del desempeño, lo cual puede dificultar la toma oportuna de decisiones estratégicas.

En cuanto al uso de indicadores de gestión, la mayoría de los participantes reportó un nivel medio o bajo de utilización. Solo una minoría indicó un uso alto, lo que indica que en general no se aprovechan sistemáticamente las métricas como herramientas de apoyo para la toma de decisiones. Esta falta de integración de indicadores puede limitar la capacidad de los directivos para anticipar problemas o medir el impacto de sus acciones.

Respecto al tiempo requerido para tomar decisiones estratégicas, se observa una tendencia a que quienes reportan un bajo uso de indicadores tardan más días en decidir, mientras que aquellos con un uso alto lo hacen en menor tiempo. Esta relación sugiere que el uso efectivo de indicadores contribuye a una toma de decisiones más ágil y posiblemente más fundamentada.

Finalmente, en la percepción sobre la calidad de las decisiones, la mayoría calificó sus decisiones como buenas, aunque se registraron respuestas negativas (mala o regular) entre quienes también reportaron menor uso de reportes o indicadores. Esto refuerza la hipótesis de que la ausencia de herramientas objetivas impacta la calidad y la eficacia del proceso decisional.

En conjunto, estos resultados permiten afirmar que la falta de indicadores de gestión influye negativamente en la eficiencia y efectividad de las decisiones estratégicas dentro de la organización. Implementar un sistema sistemático de medición y seguimiento puede ser una estrategia clave para mejorar el rendimiento global del negocio.
